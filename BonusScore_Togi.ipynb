```
numbers = []
even_count = 0
odd_count = 0

for num in range(1, 10):
  numbers.append(num)
  if num % 2 == 0: 
    even_count += 1
  else: 
    odd_count += 1

print("Jumlah angka genap: ", even_count) 
print("Jumlah anka ganjil: ", odd_count) 
```

```
datahorse['outcome'] = datahorse['outcome'].fillna((datahorse['outcome'].mode()[0]))
datahorse['surgery?'] = datahorse['surgery?'].fillna((datahorse['surgery?'].mode()[0]))
datahorse['abdomen'] = datahorse['abdomen'].fillna((datahorse['abdomen'].mode()[0]))
datahorse['rectal examination'] = datahorse['rectal examination'].fillna((datahorse['rectal examination'].mode()[0]))
datahorse['nasogastric reflux'] = datahorse['nasogastric reflux'].fillna((datahorse['nasogastric reflux'].mode()[0]))
datahorse[' nasogastric tube'] = datahorse[' nasogastric tube'].fillna((datahorse[' nasogastric tube'].mode()[0]))
datahorse['abdominal distension'] = datahorse['abdominal distension'].fillna((datahorse['abdominal distension'].mode()[0]))
datahorse['peristalsis'] = datahorse['peristalsis'].fillna((datahorse['peristalsis'].mode()[0]))
datahorse['pain'] = datahorse['pain'].fillna((datahorse['pain'].mode()[0]))
datahorse['capillary refill time'] = datahorse['capillary refill time'].fillna((datahorse['capillary refill time'].mode()[0]))
datahorse['mucus membranes'] = datahorse['mucus membranes'].fillna((datahorse['mucus membranes'].mode()[0]))
datahorse['peripheral pulse'] = datahorse['peripheral pulse'].fillna((datahorse['peripheral pulse'].mode()[0]))
datahorse['temperature of extremities'] = datahorse['temperature of extremities'].fillna((datahorse['temperature of extremities'].mode()[0]))
datahorse['total protein'] = datahorse['total protein'].fillna((datahorse['total protein'].mean()))
datahorse['packed cell volume'] = datahorse['packed cell volume'].fillna((datahorse['packed cell volume'].mean()))
datahorse['respiratory rate'] = datahorse['respiratory rate'].fillna((datahorse['respiratory rate'].mean()))
datahorse['pulse'] = datahorse['pulse'].fillna((datahorse['pulse'].mean()))
datahorse['rectal temperature'] = datahorse['rectal temperature'].fillna((datahorse['rectal temperature'].mean()))
```

```
feature = datahorse.drop(['Hospital Number','outcome'], axis=1)
label = datahorse['outcome']
```

```
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder = label_encoder.fit(label)
label_encoded = label_encoder.transform(label)
```

```
from sklearn.model_selection import train_test_split
```

```
feature_train, feature_test,label_train, label_test = train_test_split(feature, label_encoded, test_size=0.2,random_state=42)
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, precision_score, recall_score
from sklearn.model_selection import cross_val_score
```

```
#RANDOM FOREST CLASSIFIER
rf_model = RandomForestClassifier(n_estimators=500,bootstrap=True, max_features = 'sqrt', random_state=42)
rf_model.fit(feature_train, label_train)
cross_validation_scores = cross_val_score(rf_model, feature_train, label_train, cv=10)
print(cross_validation_scores, '\n')

y_pred = rf_model.predict(feature_test)
print('Accuracy of Random Forest on test set: {:.2f}'.format(rf_model.score(feature_test, label_test)))
print(f1_score(label_test, y_pred, average = 'macro'))
print(precision_score(label_test, y_pred, average = 'macro'))
print(recall_score(label_test, y_pred, average = 'macro'))
```

```
#XGBOOST
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators=500)
xgb.fit(feature_train, label_train)
cross_validation_scores = cross_val_score(xgb, feature_train, label_train, cv=10)
print(cross_validation_scores, '\n')

y_pred = xgb.predict(feature_test)
print('Accuracy of XGBoostClassifier on test set: {:.2f}'.format(xgb.score(feature_test, label_test)))
print(f1_score(label_test, y_pred, average = 'macro'))
print(precision_score(label_test, y_pred, average = 'macro'))
print(recall_score(label_test, y_pred, average = 'macro'))
```

```
#kNNclassifier
from sklearn.neighbors import KNeighborsClassifier
k_num=[*range(1, 16, 2)]
for index, val in enumerate(k_num):
  classifier_knn = KNeighborsClassifier(val, weights='distance', leaf_size=40, p=1, n_jobs=1)
  classifier_knn.fit(feature_train, label_train)
  y_pred = classifier_knn.predict(feature_test)
  print('Accuracy of kNN with k={} on test set is: {:.2f}'.format(val,classifier_knn.score(feature_test, label_test)))
  print(f1_score(label_test, y_pred, average="macro"))
  print(precision_score(label_test, y_pred, average="macro"))
  print(recall_score(label_test, y_pred, average="macro"))
```

## 2. Regression

Gunakan dataset **abalone**, dan hitung R^2, Mean Absolute Error, serta Root Mean Squared Error dari minimal 3 algoritma!

Silahkan merujuk ke file **.names** untuk melihat attributenya! Dipersilahkan untuk memproses nama feature menggunakan pandas atau mengubah langsung file .csv nya (ditambahkan header/judul per kolomnya)

```
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import warnings  
warnings.filterwarnings('ignore')
from google.colab import drive
drive.mount('/content/drive')
```

```
os.listdir('.')
```

```
dataset = pd.read_csv('abalone.csv')
dataset.head()
```

```
dataset.isnull().sum()
```

```
dataset.info()
```

```
dataset.hist(figsize=(18, 22), bins=50, xlabelsize=8, ylabelsize=8)
```

```
corr = dataset.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corr, cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':15}, cmap='Greens')
```

```
#Encoding ['Sex'] - 2 M 0F 1I
dataset['Sex'] = LabelEncoder().fit_transform(dataset['Sex'].tolist())
```

```
from sklearn.preprocessing import OneHotEncoder
transformed_sex_feature = OneHotEncoder().fit_transform(dataset['Sex'].values.reshape(-1,1)).toarray()
dataset_sex_encoded = pd.DataFrame(transformed_sex_feature, columns = ["Sex_"+str(int(i)) for i in range(transformed_sex_feature.shape[1])])
dataset = pd.concat([dataset, dataset_sex_encoded], axis=1)
```

```
dataset.head()
```

```
feature = dataset.drop(['Rings','Sex'], axis=1)
label = dataset['Rings']
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
feature_scaled = scaler.fit_transform(feature.values)
label_scaled = scaler.fit_transform(label.values.reshape(-1,1))
feature_train, feature_test,label_train, label_test = train_test_split(feature_scaled, label_scaled, test_size=0.2,random_state=42)
```

```
#RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
est_num = [*range(10, 201, 10)]
for index, val in enumerate(est_num):
    rfg = RandomForestRegressor(n_estimators = val, bootstrap = True, max_features = 'sqrt')
    rfg.fit(feature_train, label_train)
    y_pred = rfg.predict(feature_test)
    print('Accuracy of Random Forest with estimator {} on test set: {:.2f}'.format(val, rfg.score(feature_test, label_test)))
    mse = mean_squared_error(label_test, y_pred)
    mae = mean_absolute_error(label_test, y_pred)
    r2 = r2_score(label_test, y_pred)
    print("MSE (Mean Squared Error)", round(mse, 3))
    print("MAE (Mean Absolute Error)", round(mae, 3))
    print("r^2 score", round(r2,3))
    print('RMSE (Root Mean Squared Error):', round(np.sqrt(mean_squared_error(label_test, y_pred)), 3))
    print("")
```

```
#XGBoost
import xgboost as xgb
est_num = [*range(10, 201, 10)]
for index, val in enumerate(est_num):
    xgreg = xgb.XGBRegressor(objective='reg:squarederror', colsample_bytree=1, eta=0.2, learning_rate=0.15, max_depth=7, alpha=5, n_estimators = val)
    xgreg.fit(feature_train, label_train)
    y_pred = xgreg.predict(feature_test)
    print('Accuracy of xgboost with estimator {} on test set: {:.2f}'.format(val, xgreg.score(feature_test, label_test)))
    mse = mean_squared_error(label_test, y_pred)
    mae = mean_absolute_error(label_test, y_pred)
    r2 = r2_score(label_test, y_pred)
    print("MSE (Mean Squared Error)", round(mse, 3))
    print("MAE (Mean Absolute Error)", round(mae, 3))
    print("r^2 score", round(r2,3))
    print('RMSE (Root Mean Squared Error):', round(np.sqrt(mean_squared_error(label_test, y_pred)), 3))
    print("")
```

```
#SVR
from sklearn.svm import SVR

kernels = ['rbf', 'poly', 'linear']
for kernel in kernels:
    svreg = SVR(kernel=kernel)
    svreg.fit(feature_train, label_train)
    y_pred = svreg.predict(feature_test)
    print('Accuracy of SVR with kernel {} on test set: {:.2f}'.format(kernel, svreg.score(feature_test, label_test)))
    mse = mean_squared_error(label_test, y_pred)
    mae = mean_absolute_error(label_test, y_pred)
    r2 = r2_score(label_test, y_pred)
    print("MSE (Mean Squared Error)", round(mse, 3))
    print("MAE (Mean Absolute Error)", round(mae, 3))
    print("r^2 score", round(r2,3))
    print('RMSE (Root Mean Squared Error):', round(np.sqrt(mean_squared_error(label_test, y_pred)), 3))
    print("")
```

```
#kNNRegressors
from sklearn.neighbors import KNeighborsRegressor
n_neighbors = [*range(1, 20, 1)]
for index,val in enumerate(n_neighbors):
    kreg = KNeighborsRegressor(n_neighbors=val, weights='distance', algorithm='brute')
    kreg.fit(feature_train, label_train)
    y_pred = kreg.predict(feature_test)
    print('Accuracy of kNNRegressors with n_neighbor {} on test set: {:.2f}'.format(val, kreg.score(feature_test, label_test)))
    mse = mean_squared_error(label_test, y_pred)
    mae = mean_absolute_error(label_test, y_pred)
    r2 = r2_score(label_test, y_pred)
    print("MSE (Mean Squared Error)", round(mse, 3))
    print("MAE (Mean Absolute Error)", round(mae, 3))
    print("r^2 score", round(r2,3))
    print('RMSE (Root Mean Squared Error):', round(np.sqrt(mean_squared_error(label_test, y_pred)), 3))
    print("")
```
